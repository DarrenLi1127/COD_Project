```python
import os
from collections import defaultdict

def count_images_per_class(folder_path):
    # Dictionary to store class counts
    class_counts = defaultdict(int)
    
    # Iterate through files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(('.png', '.jpg', '.jpeg')):  # Add more extensions if needed
            # Split the filename and get the class name (5th element when split by '-')
            try:
                class_name = filename.split('-')[5]
                class_counts[class_name] += 1
            except IndexError:
                continue
    
    # Print results
    print("Number of images per class:")
    for class_name, count in sorted(class_counts.items()):
        print(f"{class_name}: {count}")
    
    return class_counts

# Usage example
folder_path = "/school/CSCI_2470/COD_Project/data/Train/Image"  # Replace with your folder path
counts = count_images_per_class(folder_path)
```
